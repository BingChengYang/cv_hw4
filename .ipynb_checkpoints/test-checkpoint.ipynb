```python
import torch
import torch.nn as nn
import torch.utils.data as Data
import torchvision
import torchvision.transforms as transform
from torchvision import transforms
import matplotlib
import matplotlib.pyplot as plt
import torchvision.models as models
import argparse
# my own library
import load_data_1
import net
```

```python
model = "model.pkl"
mini_batch_size = 1
scale = 3
```

```python
transform = transform.Compose([
    transform.ToTensor(),
])
```

```python
network = torch.load('./model/'+model)
print(network)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
network.to(device)
```

```python
test_set = load_data_1.Load_testdata(transform=transform)
test_load = Data.DataLoader(dataset=test_set, batch_size=mini_batch_size, shuffle=False)
```

```python
network.eval()
with torch.no_grad():
    for data in test_load:
        x = data
        x = x.to(device)
        outputs = network(x)
        print(outputs)
        outputs = outputs[0].to("cpu")
        new_img_PIL = transforms.ToPILImage()(outputs).convert('RGB')
        new_img_PIL.show() 
```
